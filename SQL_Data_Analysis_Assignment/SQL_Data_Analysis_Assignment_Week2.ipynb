# **Section A – SQL Basics**
1. Display all records


```python
pd.read_sql("SELECT * FROM superstore;", conn)
```

2. Display only Customer Name, Sales and Profit

```python
pd.read_sql("SELECT \"Customer Name\", Sales, Profit FROM superstore;", conn)
```

3. Display first 10 rows

```python
pd.read_sql("SELECT * FROM superstore LIMIT 10;", conn)
```

4. Show unique Categories

```python
pd.read_sql("SELECT DISTINCT Category FROM superstore;", conn)
```

# **Section B – Filtering**

```python
pd.read_sql("SELECT * FROM superstore WHERE State='California';", conn)
```

# **Section C – Sorting**

```python
pd.read_sql("SELECT * FROM superstore ORDER BY Sales DESC;", conn)
```

```python
pd.read_sql("SELECT * FROM superstore ORDER BY Sales DESC LIMIT 10;", conn)
```

# **Section D – Aggregate Functions**

```python
pd.read_sql("SELECT SUM(Profit) AS Total_Profit FROM superstore;", conn)
```

# **Section E – GROUP BY**

```python
pd.read_sql("SELECT Category, SUM(Sales) AS Total_Sales FROM superstore GROUP BY Category;", conn)
```

# **Section F – HAVING**

```python
pd.read_sql("SELECT Category, SUM(Sales) AS Total_Sales FROM superstore GROUP BY Category HAVING SUM(Sales)>500000;", conn)
```

# **Section I – Date Analysis**

```python
pd.read_sql("SELECT SUBSTR(\"Order Date\", -4, 4) AS Year, COUNT(*) AS Orders FROM superstore GROUP BY Year;", conn)
```

# **Section J – Advanced SQL**

```python
pd.read_sql("SELECT * FROM superstore WHERE Profit=(SELECT MAX(Profit) FROM superstore);", conn)
```
